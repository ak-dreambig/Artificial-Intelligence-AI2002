```python
import string

# Function to calculate the frequency of printable ASCII characters in a given text
def calculate_frequency(text):
    frequency = {}
    for char in text:
        if char in string.printable:
            if char in frequency:
                frequency[char] += 1
            else:
                frequency[char] = 1
    return frequency

# Function to read a cipher file and return a cipher array
def read_cipher_file(file):
    cipher_array = []
    with open(file, 'r') as f:
        for line in f:
            for char in line:
                if char in string.printable:
                    cipher_array.append(char)
    return cipher_array

# Function to encrypt a message using a given cipher array
def encrypt_message(message, cipher_array):
    encrypted_message = ''
    for char in message:
        if char in cipher_array:
            encrypted_char = cipher_array.index(char) + 32
            encrypted_message += chr(encrypted_char)
        else:
            encrypted_message += char
    return encrypted_message

# Function to decrypt a message using a given cipher array
def decrypt_message(message, cipher_array):
    
    decrypted_message = ''
    for char in message:
        if char in string.printable:
            
            char_index = cipher_array.index(char)
            
            if char_index == 0:
                decrypted_message += '\n'
            decrypted_char = chr( char_index + 31)
            decrypted_message += decrypted_char
        else:
            break
            decrypted_message += char
            
    return decrypted_message

# Function to determine the cipher for a given encrypted message
def determine_cipher(encrypted_message):
    frequency = calculate_frequency(encrypted_message)
    sorted_frequency = {k: v for k, v in sorted(frequency.items(), key=lambda item: item[1], reverse=True)}
    english_frequency = ['E', 'T', 'A', 'O', 'I', 'N', 'S', 'H', 'R', 'D', 'L', 'U', 'C', 'M', 'F', 'Y', 'W', 'G', 'P', 'B', 'V', 'K', 'X', 'Q', 'J', 'Z']
    cipher_array = [''] * 96
    for i in range(26):
        cipher_array[ord(sorted_frequency.popitem()[0]) - 32] = english_frequency[i]
    for i in range(96):
        if cipher_array[i] == '':
            cipher_array[i] = chr(i + 32)
    return cipher_array

# Example usage with provided files
cipher_array1 = read_cipher_file('cipher1.txt')
encrypted_message1 = open('encrypted1.txt', 'r').read()
decrypted_message1 = decrypt_message(encrypted_message1, cipher_array1)

print( decrypted_message1)
```
